```python
#Let us load the required libraries

import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Let us load the image

img = cv2.imread('markz.jpeg')
```

```python
plt.imshow(img)
```

Since we know that OpenCv loads an image in BGR format, so we need to convert it into RGB format to be able to display its true colors.

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

We can also convert from BGR to RGB by calling a small function.

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#Loading the classifier for frontal face

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
#We shall now detect the face by using detectMultiscale

img = cv2.imread('markz.jpeg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
#Looping the co-ordinates and drawing rectangles around them
#We will green rectangle with thickness 5
#characterizing x, y coordinates with its pixels

for (x,y,w,h) in faces:
     cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 5)
```

```python
#Converting the image to RGB and showing the image

plt.imshow(convertToRGB(img))
```

```python
#Saving the image

cv2.imwrite("markz1.png", faces)
```
